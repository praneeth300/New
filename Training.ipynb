```python
x=df.drop(["Price","optimal_hours"],axis=1)
y1=df["Price"]
y2=df["optimal_hours"]

import xgboost as xg
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y2,test_size=0.2,random_state=0)

scale_cols=["Dept_date","duration","dept_hours"]

from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
def scale(train,test,cols):
    mm.fit(train[cols].values.reshape(-1,1))
    train[cols]=mm.transform(train[cols].values.reshape(-1,1))
    test[cols]=mm.transform(test[cols].values.reshape(-1,1))

    return train[cols] ,test[cols]

for i in scale_cols:
    x_train[i],x_test[i]=scale(x_train,x_test,i) 


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
params={'max_depth':[5,10,15,20,50],
        'min_samples_split':[2,3,5,10]}
rfg_time=RandomForestClassifier(n_estimators=50)
cv_time=RandomizedSearchCV(rfg_time,params,cv=5,n_jobs=-1,verbose=2)
cv_time.fit(x_train,y_train)
```

```python
print(cv_time.best_params_)
model_time=cv_time.best_estimator_
model_time.fit(x_train,y_train)

pred = model_time.predict(x_test)

from sklearn.metrics import classification_report

print(classification_report(y_test,pred))

xxx = pd.DataFrame({'Actual time': y_test, 'Predicted time': pred})
xxx.head(5)
```

```python
# price
x=df.drop(["Price","dept_hours"],axis=1)
y1=df["Price"]


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y1,test_size=0.2,random_state=0)

scale_cols=["Dept_date","duration","optimal_hours"]

from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
def scale(train,test,cols):
    mm.fit(train[cols].values.reshape(-1,1))
    train[cols]=mm.transform(train[cols].values.reshape(-1,1))
    test[cols]=mm.transform(test[cols].values.reshape(-1,1))

    return train[cols] ,test[cols]

for i in scale_cols:
    x_train[i],x_test[i]=scale(x_train,x_test,i) 

from sklearn.ensemble import RandomForestRegressor
params={'max_depth':[5,10,15,20,50],
        'min_samples_split':[2,3,5,10]}
rfg_price=RandomForestRegressor(n_estimators=50)
cv_price=RandomizedSearchCV(rfg_price,params,cv=5,n_jobs=-1,verbose=2)
cv_price.fit(x_train,y_train)
```

```python
print(cv_price.best_params_)
model_price=cv_price.best_estimator_
model_price.fit(x_train,y_train)

pred = model_price.predict(x_test)
from sklearn.metrics import mean_squared_error as mse , r2_score
from math import sqrt
rms = sqrt(mse(y_test, pred))
score = r2_score(y_test, pred)
print(rms)
print(score)

xxx = pd.DataFrame({'Actual price': y_test, 'Predicted price': pred})
xxx.head(5)
```
